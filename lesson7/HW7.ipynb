```python
class Animal:
    
    def __init__(self):
        self._colour = "" #инкапсуляция
    
    #@property
    def get_colour(self):
        return self._colour
    
    #@colour.setter
    def set_colour(self, colour):
        self._colour = colour
        
    def sound(self):
        print("*sound*")

class Cat(Animal): #наследование
    
    def sound(self): #полиморфизм
        print("meow")
        
dog = Animal()
dog.set_colour("orange")
dog.get_colour()

tiger = Cat()
tiger.set_colour("black")
tiger.get_colour()
tiger.sound()
```

```python
class Money:
    
    def __init__(self, rubles, cents):
        self.rubles = rubles
        self.cents = cents
    
    def print_money(self):
        if self.cents > 10:
            print(self.rubles,self.cents, sep = ',')
        else:
            print(f"{self.rubles},0{self.cents}")
        
    def sum_money(self, rubles, cents):
        if (self.cents + cents) > 99:
            self.cents = (self.cents + cents) % 100
            self.rubles += (rubles + 1)
        else:
            self.cents += cents
            self.rubles += rubles
        self.print_money()
        return self.rubles + self.cents / 100
    
    def sub_money(self, rubles, cents):
        if cents > self.cents:
            self.rubles -= (rubles + 1)
            if (self.rubles < 0):
                self.rubles += 1
                self.cents = cents - self.cents
            else:
                self.cents = 100 - (cents - self.cents)
        else:
            self.cents -= cents
            self.rubles -= rubles
        self.print_money()
        return self.rubles + self.cents / 100
        
    def mul_money(self, num):
        return (self.rubles + self.cents / 100) * num
    
    def div_money(self, num):
        return (self.rubles + self.cents / 100) / num
    
    def compare(self, num):
        money = self.rubles + self.cents / 100
        if money > num:
            print("Savings are greater than num")
        elif money < num:
            print("Savings are lesser than num")
        else:
            print("Savings are equal to num")
        
wallet = Money(15,64)
wallet.print_money()
wallet.sum_money(16,56)
#wallet.sub_money(33,21)
print(wallet.mul_money(1.5))
```

```python
class Matrix:
    
    def __init__(self, n, m):
        self.rows = n
        self.columns = m
        self.matrix = []
    
    def initialize(self):
        for i in range(self.rows):
            temp = []
            for j in range(self.columns):
                element = int(input())
                temp.append(element)
            self.matrix.append(temp)
        return(self.matrix)
    
    def print_matrix(self):
        for i in range(0, self.rows):
            for j in range(0, self.columns):
                print(self.matrix[i][j], end = " ")
            print("\n")
            
    def sum_matrix(self, obj):
        if self.rows != obj.rows or self.columns != obj.columns:
            return "Error, matrixes must be the same size"
        
        result_matrix = []
        
        for i in range(self.rows):
            temp = []
            for j in range(self.columns):
                temp.append((self.matrix[i][j] + obj.matrix[i][j]))
            result_matrix.append(temp)
        return result_matrix
    
    def mul_matrix_by_number(self, num):
        result_matrix = []
        
        for i in range(self.rows):
            temp = []
            for j in range(self.columns):
                temp.append((self.matrix[i][j] * num))
            result_matrix.append(temp)
        return result_matrix

abc = Matrix(3,3)
abc.initialize()
abc.print_matrix()

cba = Matrix(3,3)
cba.initialize()
cba.print_matrix()

print(abc.sum_matrix(cba))
cba.mul_matrix_by_number(2)
```
