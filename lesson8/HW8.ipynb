```python
import matplotlib.pyplot as plt
```

```python
img1 = plt.imread(f"pictures/1.jpeg") # dataset: https://www.kaggle.com/datasets/alessiocorrado99/animals10/data
plt.imshow(img1) # чтение картинки
img1 # массив значений каналов внутри картинки
```

```python
img1.shape # размер картинки
```

```python
invert = img1.copy() # копирование изображения в отдельную переменную
invert = 255 - img1 # инвертирование цветов изображения
plt.imshow(invert)
invert
```

```python
fig, ax = plt.subplots(7, 2, figsize=(8, 7 * 3)) # 7 - количество изображений (строк), 2 - колво столбцов
ax[0,0].set_title("Original img") # название левого столбца
ax[0,1].set_title("Reveresed img") # название правого столбца

for i in range(7):
    try:
        img = plt.imread(f"pictures/{i+1}.jpeg") # загрузка изображения
    except FileNotFoundError: # если формат файла не jpeg
        img = plt.imread(f"pictures/{i+1}.jpg") 
    invert = img.copy()
    invert = 255 - img
    
    ax[i,0].imshow(img)
    ax[i,0].axis('off') # выключение осей, т.к. они не нужны

    ax[i,1].imshow(invert)
    ax[i,1].axis('off')
```
