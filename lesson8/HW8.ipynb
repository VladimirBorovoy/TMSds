```python
fig, ax = plt.subplots(7, 2, figsize=(8, 7 * 3)) # 7 -   (), 2 -  
ax[0,0].set_title("Original img") #   
ax[0,1].set_title("Reveresed img") #   

for i in range(7):
    try:
        img = plt.imread(f"pictures/{i+1}.jpeg") #  
    except FileNotFoundError: #     jpeg
        img = plt.imread(f"pictures/{i+1}.jpg") 
    invert = img.copy()
    invert = 255 - img
    
    ax[i,0].imshow(img)
    ax[i,0].axis('off') #  , ..   

    ax[i,1].imshow(invert)
    ax[i,1].axis('off')
```
