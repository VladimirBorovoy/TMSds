<h1>Представим круговые диаграммы для топ-50 университетов каждого года по странам<h1>

```python
# Подсчет количества каждой страны
country_counts = top_50_2011['country'].value_counts()

# Построение круговой диаграммы
plt.figure(figsize=(12, 8))
plt.pie(country_counts, labels=country_counts.index, autopct='%1.1f%%', startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Количество университетов в топ-50 по странам для 2011 года')
plt.show()
```
