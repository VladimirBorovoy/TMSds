```python
def my_matrix_sum(A: list, B: list) -> list:
    result: list = []
    if len(A) == len(B) and len(A[0]) == len(B[0]):
        for i in range(len(A)):
            temp = []
            for j in range(len(A[0])):
                temp.append(A[i][j] + B[i][j])
            result.append(temp)
        return result
    else:
        print("Matrix must be same size")
        return result
    
def my_matrix_mul(A: list, B: list) -> list:
    result: list = []
    if len(A[0]) == len(B):
        res = [[sum(a * b for a, b in zip(A_row, B_col)) 
                for B_col in zip(*B)] # * -    
                        for A_row in A]
        result.append(res)
        return result
    else:
        print("Matrix A must have the same number of columns as matrix's B number of rows")
        return result
    
A4 = [[1, -2, 8], [7, 5, -2], [-2, 3, 0]]
B4 = [[0, 2, 4], [1, -3, -2], [-2, 1, 3]]
C4 = [[0, 3], [5, -4], [7, -1]]

A4np = np.array([[1, -2, 8], [7, 5, -2], [-2, 3, 0]])
B4np = np.array([[0, 2, 4], [1, -3, -2], [-2, 1, 3]])
C4np = np.array([[0, 3], [5, -4], [7, -1]])
print(f"My plus: {my_matrix_sum(A2, B2)}")
print(f"Np plus: \n {A2np + B2np}")

print(f"My plus: {my_matrix_mul(A2, B2)}")
print(f"Np plus: \n {np.dot(A2, B2)}")
```

```python
import matplotlib.pyplot as plt

def svertka(image, kernel):
    kernel_height, kernel_width = kernel.shape
    image_height, image_width, image_channels = image.shape
    
    output = np.zeros((image_height - kernel_height + 1, image_width - kernel_width + 1, image_channels))
    
    for channel in range(image_channels):
        for x in range(image_height - kernel_height + 1):
            for y in range(image_width - kernel_width + 1):
                output[y, x, channel] = np.sum(
                    image[y:y + kernel_height, x:x + kernel_width, channel] * kernel
                ).clip(0, 255)

    return output.astype(np.uint8)


kernel = np.array([[-1, -1, -1],
                   [-1, 9, -1],
                   [-1, -1, -1]])


image = plt.imread('2.jpg')
print(image.shape)
print(image)
plt.imshow(image)

new_img = svertka(image, kernel)

plt.imshow(new_img)
```
